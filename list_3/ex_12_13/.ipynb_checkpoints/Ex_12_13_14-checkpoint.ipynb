## Dataset preparation

### Get rid of string values

```python
def create_dummy_variable(dataframe, column):
    try:
        dummy_encoding = pd.get_dummies(dataframe[column], prefix=column)
        dataframe.drop(column, axis=1, inplace=True)
        dataframe = pd.concat([dataframe, dummy_encoding], axis=1)
    except KeyError as e:
        print(f"Column '{column}' not found!")
    return dataframe
```

```python
df = create_dummy_variable(df, 'Married')
df = create_dummy_variable(df, 'Ethnicity')
df = create_dummy_variable(df, 'Gender')
df = create_dummy_variable(df, 'Student')
```

```python
df
```

### Get rid of dummy variable trap
After few tests it appears that dropping the column regarding ethnicity results in slightly worse accuracies.

```python
df.drop('Gender_Male', axis=1, inplace=True)
df.drop('Married_Yes', axis=1, inplace=True)
# df.drop('Ethnicity_African American', axis=1, inplace=True) # decreases accuracy!
df.drop('Student_Yes', axis=1, inplace=True)
```

```python
df_a = df.copy(deep=True)
df_b = df.copy(deep=True)
```

```python
def get_train_test_split(dataframe, features, results):
    X = df_a[features]
    y = df_a[results]
    return train_test_split(X, y, test_size=0.2, random_state=13) # X_train, X_test, y_train, y_test
```

### a) if a given person has an income greater than 50 (hint: create new indicator variable)

```python
df_a['Income > 50'] = df_a.apply(lambda row: 1 if row['Income'] > 50 else 0, axis=1)
```

```python
df_a
```

```python
features = list(df_a.columns)
features.remove('Income > 50')
result = ['Income > 50']
X_a, y_a = df_a[features], df_a[result]

X_a_train, X_a_test, y_a_train, y_a_test = get_train_test_split(df_a, features, result)
```

#### sklearn

```python
model = LogisticRegression(C=1e9, fit_intercept=False, solver='liblinear', penalty='l2', max_iter=10000)
mdl = model.fit(X_a_train, y_a_train.to_numpy().ravel())
test_accuracy = model.score(X_a_test, y_a_test.to_numpy().ravel())
print(test_accuracy)
```

## Grid Search with Cross Validation

```python
def perform_grid_search_cv(model, parameters, X, y, cv=10, verbose=10, n_jobs=-1):
    grid_object = GridSearchCV(model, parameters, scoring='accuracy',
                               cv=cv, verbose=verbose, n_jobs=n_jobs, iid=True)
    grid_object.fit(X, y)
    return grid_object
```

### Dataset a)
#### K-Nearest Neighbour

```python
params = {'n_neighbors': [1,2,4,8,20,50]}
knn_grid_object = perform_grid_search_cv(KNeighborsClassifier(), params, X_a, y_a.to_numpy().ravel())

best_knn_score = knn_grid_object.best_score_
knn_results = knn_grid_object.cv_results_
best_knn_estimator = knn_grid_object.best_estimator_
```

```python
print(f'Best score = {best_knn_score}')
```

#### Decision Tree

```python
params = {'max_depth': [1, 5, 10, 100], } 
dt_grid_object = perform_grid_search_cv(DecisionTreeClassifier(), params, X_a, y_a.to_numpy().ravel())

best_dt_score = dt_grid_object.best_score_
dt_results = dt_grid_object.cv_results_
best_dt_estimator = dt_grid_object.best_estimator_
```
